```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(2010,2020)
y = np.array([20,40,32,61,82,70,90,130,160,210])

rev = []
for i in y[::-1]:
    rev.append(i)
ry = np.array(rev)
ry
```

```python
??plt.plot
```

```python
plt.plot(x,y,'black',label='raise',ls=':')
plt.xlabel('Year')
plt.ylabel('No of units')
plt.title('SALES CHART OF TV')
plt.legend(loc=4)
```

```python
fig = plt.figure(figsize=(7,5))

g1 = fig.add_axes([0.1,0.1,0.9,0.9])
g2 = fig.add_axes([0.2,0.55,0.4,0.4])

g1.plot(x,y,'black',ls=':',lw='2', marker='s',markersize='7',markerfacecolor='green',label='Raise')
g2.plot(x,ry,'black',ls=':',lw='2',marker='s',markersize='5',markerfacecolor='red',label='Fall') 
# for marker, linestyle, color run ---> ??plt.plot

g1.set_xlabel('Year')
g1.set_ylabel('Units of TV')
g1.set_title('SALES CHART TV Vs RADIO')

g2.set_xlabel('Year')
g2.set_ylabel('Units of Radio')

g1.legend(loc=4)
g2.legend()
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,88,77,82,112]
blood_sugar_women = [67,98,85,120,133,150,84,69,89,79,120,112,100]


plt.hist([blood_sugar_men,blood_sugar_women],
         bins=[80,100,120,150],
         rwidth=0.9,
         color=['green','orange'],
         label=['men','women'],
         orientation='horizontal')

plt.xlabel('Blood sugar level')
plt.ylabel('Number of patients')
plt.title('Blood sugar analysis')
plt.legend()
```
