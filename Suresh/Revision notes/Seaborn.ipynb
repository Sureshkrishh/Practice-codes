```python
sns.distplot(df['Attack'])
plt.show()
```

```python
??sns.kdeplot
```

```python
sns.set(color_codes=True)

kd = sns.kdeplot(df['Attack'],color='black',shade=True)

kd.set_title('kernal density plot')
kd.set_xlabel('Attack point')
kd.set_ylabel('Attack probability')
plt.show()
```

```python
??sns.rugplot
```

```python
sns.kdeplot(df['Attack'],
            df['Defense'],
            color='g',shade=True)

sns.rugplot(df['Attack'],
            height=0.1,
            color='red')

sns.rugplot(df['Defense'],
            height=0.1,
            color='black',
            vertical=True)

plt.xlim(0,155)
plt.ylim(0,200)
plt.show()
```

```python
al = sns.distplot(df['Attack'],
             hist = True,
             kde = True,
             rug = True,
             kde_kws = {'color':'black','label':'density','marker':'s','markersize':2,'markerfacecolor':'white'},
             hist_kws = {'color':'green','histtype':'step','lw':3,'label':'histo'},
             rug_kws = {'color':'red','height':0.2,'label':'points'})

al.set_xlabel('attack points')
al.set_ylabel('frequency')
al.set_title('distribution')
al.legend()
```

```python
rg = sns.rugplot(df['Attack'],
                axis = 'x',
                height=0.1,
                color='red')

rg = sns.rugplot(df['Defense'],
                axis = 'y',
                height=0.1,
                color='green')
```

```python
??sns.jointplot
```

```python
sns.set(color_codes=True)

sns.jointplot(x='Attack',
              y='Defense',
              data=df,
              xlim={0,150},
              ylim={0,200},
              kind='kde', #other kinds - scatter,hex,reg,resid
              color='black')

plt.show()
```

```python
??sns.heatmap
```

```python
plt.subplots(figsize=(8,8))
sns.heatmap(df.corr(),            
            vmax=0.8,
            square=True,
            annot=True,
            fmt='.3f',
            cmap='Purples', # plt.colormaps() - gives all cmap available
            mask=False)

plt.show()
```

```python
??sns.lmplot
```

```python
sns.lmplot(x='Attack',y='Defense',
           data=df)
plt.show()
```

```python
pn = sns.load_dataset('penguins')
pn.head()
```

```python
print(pn['species'].unique())
print(pn['island'].unique())
print(pn['body_mass_g'].mean())
```

```python
??sns.pairplot
```

```python
sns.pairplot(pn)
```

```python
sns.pairplot(pn,
             x_vars=['bill_depth_mm','bill_length_mm','flipper_length_mm'],
             y_vars=['body_mass_g'],
             kind='scatter',
             hue='species',
             height=5,
             aspect=0.8, diag_kind=None)

plt.show()
```

```python
sns.pairplot(pn,
             x_vars=['bill_depth_mm','bill_length_mm','flipper_length_mm'],
             y_vars=['body_mass_g'],
             kind='scatter',
             hue='island',
             height=5,
             aspect=0.8)

plt.show()
```

```python
??sns.stripplot
```

```python
plt.subplots(figsize=(10,8))

sns.stripplot(x='island',
              y='flipper_length_mm',
              data=pn,
              size=5,
              hue='sex',
              color='violet',
              jitter=0.3,  # greater the jitter value higher the gap between distribution
              linewidth=1) # adds border to the markers 
```

```python
??sns.swarmplot
```

```python
plt.subplots(figsize=(10,8))

sns.swarmplot(data=pn,
             x='island',
             y='body_mass_g',
             hue='sex')
```

```python
??sns.boxplot
```

```python
plt.subplots(figsize=(15,8))

sns.boxplot(data=pn,
            x='sex',
            y='body_mass_g',
            hue='island')

plt.xlabel('SEX')
plt.ylabel('BODY MASS')
plt.title('Statistical inference')
plt.legend()
```

```python
??sns.violinplot
```

```python
plt.subplots(figsize=(15,8))

sns.violinplot(data=pn,
               x='island',
               y='flipper_length_mm',
               hue='sex',
               pallete='muted',
               split=False) # try split = True to get data in single kde

plt.xlabel('ISLAND')
plt.ylabel('FLIPPER LENGTH')
plt.title('Statistical inference')
plt.legend()
```
