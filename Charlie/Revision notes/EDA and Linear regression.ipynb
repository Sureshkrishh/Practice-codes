### Missing value treatment

```python
# Missing values in Time_taken 
# Since its quantitative data we can replace the missing value with mean

df.Time_taken = df.Time_taken.fillna(df.Time_taken.mean()) 
```

```python
# Missing values in Genre
# Since its a cateegorical data we can replace the missing value with most occured category

sns.countplot(x='Genre', data=df) # Thiller is the most frequently occured categorical data
df.Genre = df.Genre.fillna('Thriller')
```

```python
print(df.Time_taken.isnull().sum())
print(df.Genre.isnull().sum())
```

```python
df.Marketin_expense = np.where(df.Marketin_expense>df.Marketin_expense.quantile(0.99),
                               df.Marketin_expense.quantile(0.99),
                               df.Marketin_expense)

df.Time_taken = np.where(df.Time_taken<df.Time_taken.quantile(0.01),
                        df.Time_taken.quantile(0.01),
                        df.Time_taken)

df.Twitter_hastags = np.where(df.Twitter_hastags>df.Twitter_hastags.quantile(0.99),
                               df.Twitter_hastags.quantile(0.99),
                               df.Twitter_hastags)

df.Avg_age_actors = np.where(df.Avg_age_actors<df.Avg_age_actors.quantile(0.01),
                        df.Avg_age_actors.quantile(0.01),
                        df.Avg_age_actors)
```

# LINEAR REGRESSION

```python
# creating the target variable as the last column

trans_df['Target'] = trans_df['Collection']
lm_df = trans_df.drop('Collection',axis=1)
lm_df.head()
```

```python
# Creating independent (X) and dependent (Y) dataset
X = lm_df.drop('Target',axis=1)
y = lm_df['Target']

# Checking the shape of the dataset
print(X.shape)
print(y.shape)
```

```python
# Checking statistical significance

import statsmodels.api as sm
mod = sm.OLS(y,X)
res = mod.fit()
print(res.summary())
```

```python
# Observation from the model summary

    # For one unit rise in independent variable, the target variable increases/decreases by coef times.
    
    # R-squared and adjusted R-squared value is 0.95:
        # which means 95% variance in the outcome variable is explained by the model.
        
    # For the two tailed test P>|t| is greater than 0.025 for Critic_rating and Time_taken
        # in other words, value of t lies between [0.025 and 0.975] (fail to reject region), so the null hypothesis is true
        # which means these variables are staistically insignificant and can be removed.
```

```python
# Creating dataframe with statistically significant variable

sig_vars = ['Marketin_expense', 'Production_expense', 'Multiplex_coverage',
            'Budget', 'Movie_length', 'Trailer_views', '3D_available_YES']
X_vars = X[sig_vars]
X_vars.head()
```

```python
# Importing required modules from scikit learn library
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## 70:30 Split Training and testing

```python
# Splitting the data for training and testing (70:30 split)
X_train, X_test, y_train, y_test = train_test_split(X_vars, y, test_size=0.3, random_state=42) #random_state can be any integer

print('70% of indepenedent variable training data is',X_train.shape)
print('30% of indepenedent variable testing data is',X_test.shape)
print('70% of depenedent variable training data is',y_train.shape)
print('30% of depenedent variable testing data is',y_test.shape)
```

```python
# Evaluating R-square and model intercept (value of b0 in y'=b0+b1x) for train-test split
linreg = LinearRegression()
model = linreg.fit(X_train,y_train)
print('R-squared value for training dataset is %.2f' %model.score(X_train,y_train))
print('R-squared value for testing dataset is %.2f' %model.score(X_test,y_test))
print('The estimated model intercept is %.2f' %model.intercept_)
```

```python
# coefficient of all independet variable

coeff_df = pd.DataFrame(X_vars.columns,columns=['Independent variable'])
coeff_df['Coefficient'] = pd.Series(model.coef_)
coeff_df
```

```python
# Prediction on training and testing datasets
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
pred_all = model.predict(X_vars)

# Calculating Mean Squared Error (MSE)
print('MSE of training dataset: %.2f' %np.mean((y_train - pred_train)**2))
print('MSE of testing dataset: %.2f' %np.mean((y_test-pred_test)**2))
print('MSE of overall dataset: %.2f' %np.mean((y-pred_all)**2))
```

```python
# PLoting scatter plot for residuals in train and test to see if there is any pattern
train = plt.scatter(pred_train,(y_train-pred_train),color='blue',alpha=0.5,label='TRAINING')
test = plt.scatter(pred_test,(y_test-pred_test),color='red',alpha=0.5,label="TESTING")
plt.hlines(y=0,xmin=-10,xmax=100000)
plt.title('RESIDUAL PLOT')
plt.xlabel('Predicted points')
plt.ylabel('Residual points')
plt.legend(loc=3)
```

```python
# There is no pattern in residual so the model is correctly captured
```

```python
# Creating Predected_target Ranking comparision of target and predicted variables
lm_df['Predicted_target'] = pd.DataFrame(pred_all).round(decimals=2)
lm_df['Target_rank'] = pd.qcut(lm_df['Target'].rank(method='first').values,5,duplicates='drop').codes+1
lm_df['Predicted_rank'] = pd.qcut(lm_df['Predicted_target'].rank(method='first').values,5,duplicates='drop').codes+1
lm_df.head()
```
